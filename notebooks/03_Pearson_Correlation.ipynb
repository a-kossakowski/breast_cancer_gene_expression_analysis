```python
# Importing necessery modules

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the cleaned, normalized and filtered data
df = pd.read_csv('../data/filtered_normalized_gene_expression_data.csv')
```

## Non-Filtered Heatmap: Understanding Overall Relationships

The non-filtered heatmap provides a comprehensive view of the relationships between all samples (genes) in the dataset. Each cell in the heatmap represents the Pearson correlation coefficient between two samples, ranging from -1 to 1. A value close to 1 indicates a strong positive correlation, a value close to -1 indicates a strong negative correlation, and a value around 0 indicates no correlation.

The purpose of this heatmap is to give an initial, broad understanding of the dataset. It helps us identify patterns, trends, and potential outliers at a glance. However, because it includes all correlations, it can be overwhelming and may include noise or less relevant relationships.

```python
# Calculate the correlation matrix
correlation_matrix = df.iloc[:, 0:].drop(['ENSG', 'HGNC'], axis=1).astype(float).corr()

# Create a heatmap
plt.figure(figsize=(20, 20))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Heatmap of Correlation Matrix")
plt.xlabel("Samples")
plt.ylabel("Samples")
plt.show()

# Save the coorelation matrix
correlation_matrix.to_csv('../data/correlation_matrix.csv')
```

## Filtered Heatmap: Focusing on Strong Correlations

The filtered heatmap narrows down the focus to only strong correlations, in this case, those with an absolute value greater than 0.9. By doing this, we remove the "noise" and less relevant relationships, making it easier to identify and interpret meaningful correlations between samples (or genes).

This heatmap is particularly useful for pinpointing specific areas of interest for further analysis. For example, a strong positive correlation between two samples could suggest similar gene expression profiles, which might be indicative of similar biological processes or conditions. On the other hand, a strong negative correlation might indicate opposing biological activities.

By focusing on these strong correlations, I can more effectively identify relationships that are likely to be significant and worthy of further investigation.

```python
# Filter the correlation matrix to only include strong correlations
filtered_corr_matrix = correlation_matrix[(correlation_matrix.abs() > 0.9) & (correlation_matrix != 1)]

# Create a heatmap of the filtered correlation matrix
plt.figure(figsize=(20, 20))
sns.heatmap(filtered_corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Filtered Heatmap of Correlation Matrix")
plt.xlabel("Samples")
plt.ylabel("Samples")
plt.show()

# Save new filtered correlation matrix into separate file
filtered_corr_matrix.to_csv('../data/filtered_corr_matrix.csv')
```
