# Exploratory Data Analysis (EDA) for Gene Expression Data
This notebook aims to explore the structure and patterns in the cleaned, normalized, and filtered gene expression data.

```python
# Importing necessery modules

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the cleaned, normalized and filtered data
df = pd.read_csv('../data/filtered_normalized_gene_expression_data.csv')
```

```python
# Show the first few rows of the data
df.head(3)
```

```python
# Generate summary statistics
pd.set_option('display.max_columns', None)
summary_stats = df.describe()
summary_stats
```

## Summary Statistics

Count: All samples have a count of 21,883, indicating that there are no missing values in the dataset. This is crucial for ensuring the reliability of any subsequent analyses.

Mean: The average gene expression levels for each sample range from around 39 to 45. This suggests that the samples have similar average expression levels, which is a good sign for comparability.

Standard Deviation (std): The standard deviation varies significantly across samples. This indicates different levels of variability in gene expression among the samples, which could be biologically meaningful or could indicate noise.

Min: All samples have a minimum value of 0, which is expected in normalized gene expression data.

25% (Q1): The 25th percentile varies across samples but is generally low. This indicates that a quarter of the genes in each sample have low expression levels, which is typical in transcriptomic data.

50% (Median): The median values are generally around 5 to 7, much lower than the mean. This suggests that the data may be right-skewed, with a few genes showing very high expression levels.

75% (Q3): The 75th percentile varies across samples but is generally around 25 to 45. This indicates that three-quarters of the genes have moderate expression levels.

Max: The maximum values are extremely high compared to the 75th percentile. This indicates the presence of outliers or genes with very high expression levels, which could be of particular interest for further investigation.

## Boxplot Construction
A boxplot is an excellent tool for visualizing the distribution of gene expression levels across different samples. It provides a five-number summary of the data ("minimum," "Q1," "median," "Q3," and "maximum") and can help identify outliers. The boxplot can also give insights into the symmetry of the data, the tightness of the data grouping, and potential skewness. Given the wide range of expression levels and the presence of outliers, as indicated by the summary statistics, a boxplot can offer a quick, informative look at the data's distribution.

```python
# Create boxplots to visualize the distribution of gene expression values across different samples
plt.figure(figsize=(20, 10))
sns.boxplot(data=df.iloc[:, 0:].drop(['ENSG', 'HGNC'], axis=1).astype(float))
plt.title("Boxplot of Gene Expression across Samples")
plt.xlabel("Samples")
plt.ylabel("Gene Expression")
plt.xticks(rotation=70)
plt.tight_layout()
plt.show()
```
